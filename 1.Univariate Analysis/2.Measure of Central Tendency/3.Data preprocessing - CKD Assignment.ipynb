```python
'''
import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[Quan])
df=imp.transform(dataset[Quan])
df=pd.DataFrame(df,columns=Quan)
'''

import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy='mean')
imp.fit(dataset[Quan])
quandataset=imp.transform(dataset[Quan])
quandataset=pd.DataFrame(quandataset,columns=Quan)
```

```python
quandataset
```

```python
quandataset.isnull().sum()
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imp.fit(dataset[Qual])
qualdataset=imp.transform(dataset[Qual])
qualdataset=pd.DataFrame(qualdataset,columns=Qual)
```

```python
qualdataset
```

```python
qualdataset.isnull().sum()
```

```python
frames = [quandataset, qualdataset]

Finaldataset = pd.concat(frames,axis=1)
```
